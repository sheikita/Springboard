## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties')
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

# JSON - Q3.Solution
Filling the missing names at for variable 'mjtheme_namecode'

```python
#Normalize variable 'mjtheme_namecode' and assign it nor_fill
nor_fill = json_normalize(j_data, 'mjtheme_namecode')
```

```python
#Create a dict to map code to name 
code_dict = {}

for row in nor_fill.itertuples():
    if row[2] != '':
        code_dict[row[1]] = row[2]
```

```python
#Verify that codes are mapped
print(code_dict)
```

```python
#Fill the missing values 

for row in nor_fill.itertuples():
    if row[2] == '':
        nor_fill.get(row[0], 'name', code_dict[row[1]])
```

```python
#Validate assignment of missing value for code 11
nor_fill.head()
```

```python
print('Top 10 themes')
nor_fill.name.value_counts().head(10)
```
